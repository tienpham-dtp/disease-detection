```python
import pandas as pd
import numpy as np
import math

```

```python
df_train = pd.read_csv('challengeTrainLabeled.csv')
```

```python
def nan_replace3(x):
    if math.isnan(x) == True:
        return 3
    else:
        return x
    
def nan_replace7(x):
    if math.isnan(x) == True:
        return 7
    else:
        return x
    
def nan_replace9(x):
    if math.isnan(x) == True:
        return 9
    else:
        return x

def nan_replace99(x):
    if math.isnan(x) == True:
        return 99
    else:
        return x

```

```python
def preprocessing(df_train): 
    
    #drop irrelevant columns:
    df_train = df_train.drop(columns = ['Exam_period','Country','Language','Self_respond_sp','Self_respond2_sp',
                                       'Interview_lang','Self_respond_fam','Self_respond2_fam','HH_gender','HH_gender',
                                       'HH_country','fluoride_water_level','fluoride_water_level'
                                        ,'fluoride_plasma_level','Num_household','dentition_status'])
    df_train = df_train.drop(columns = 'last_alcohol')
    df_train = df_train.drop(columns = ['Unnamed: 0','Release_Cycle','Status'])
    
    #recent_alcohol values
    df_train['recent_alcohol'].fillna(-1, inplace = True)

    #we notice that Income_household is the same as Income_family - can be implied interchangably 
    #Replace NaNs in Income_family with Income_household
    for i in range(0,df_train.shape[0]):
        if (math.isnan(df_train.iloc[i,df_train.columns.get_loc("Income_household")]) == False) & (math.isnan(df_train.iloc[i,df_train.columns.get_loc("Income_family")])== True):
            df_train.iloc[i,df_train.columns.get_loc("Income_family")] = df_train.iloc[i,df_train.columns.get_loc("Income_household")]
    
    #create a temp_df to calculate the poverty threshold for each family group
    temp_df = df_train[['Income_family','Num_family','Income_ratio']].copy()
    temp_df['poverty_threshold'] = temp_df['Income_family']/temp_df['Income_ratio']
    temp_df = temp_df[~temp_df['poverty_threshold'].isin([np.inf,-np.inf])]

    #we choose median because it is not as skewed as mean()
    poverty_lines = list(temp_df.groupby('Num_family')['poverty_threshold'].median())


    #fill NaNs for Income_ratio + we checked that there is no NaNs in Num_family
    for i in range(0, df_train.shape[0]):
        if pd.isnull(df_train.iloc[i,df_train.columns.get_loc('Income_ratio')]) == True:
            df_train.iloc[i,df_train.columns.get_loc('Income_ratio')] = (df_train.iloc[i,df_train.columns.get_loc('Income_family')])/(poverty_lines[int(df_train.iloc[i,df_train.columns.get_loc('Num_family')]-1)]) 
     
    #for cases where both Income_ratio and Income_family are both null - impute with (-1)
    for i in range(0,df_train.shape[0]):
        if ((pd.isnull(df_train.iloc[i,df_train.columns.get_loc('Income_ratio')]) == True) & 
           pd.isnull(df_train.iloc[i,df_train.columns.get_loc('Income_family')]) == True):
            df_train.iloc[i,df_train.columns.get_loc('Income_ratio')] = -1
            df_train.iloc[i,df_train.columns.get_loc('Income_family')] = -1
    
    df_train = df_train.drop(columns = ['Income_household'])
    #pregnancy_test - replace values
    for i in range(0, df_train.shape[0]):
        if pd.isnull(df_train.iloc[i,df_train.columns.get_loc('pregnancy_test')]) == True:
            if df_train.iloc[i,df_train.columns.get_loc('Gender')] == 2:
                df_train.iloc[i,df_train.columns.get_loc('pregnancy_test')] =  3
            elif df_train.iloc[i,df_train.columns.get_loc('Gender')] == 1:
                df_train.iloc[i,df_train.columns.get_loc('pregnancy_test')] =  4
    
    
    #if cells in 'Race1' and 'Race2' don't match - replace with '5'
    for i in range(0, df_train.shape[0]):
        if (df_train.iloc[i,df_train.columns.get_loc('Race1')] != df_train.iloc[i,df_train.columns.get_loc('Race2')]):
            df_train.iloc[i,df_train.columns.get_loc('Race1')] = 5

    #fillna for US_citizen
    df_train = df_train.fillna({'US_citizen':9})

    Q1_index = df_train.columns.get_loc('Q1')
    Q16_index = df_train.columns.get_loc('Q16')
    Q17_index = df_train.columns.get_loc('Q17')
    Q27_index = df_train.columns.get_loc('Q27')


    #replace all Q1-Q26 with unknown values
    df_train['Q1'] = df_train['Q1'].apply(nan_replace7)
    df_train['Q2'] = df_train['Q2'].apply(nan_replace7)
    df_train['Q3'] = df_train['Q3'].apply(nan_replace7)
    df_train['Q4'] = df_train['Q4'].apply(nan_replace7)
    df_train['Q5'] = df_train['Q5'].apply(nan_replace7)
    df_train['Q6'] = df_train['Q6'].apply(nan_replace7)
    df_train['Q7'] = df_train['Q7'].apply(nan_replace7)
    df_train['Q8'] = df_train['Q8'].apply(nan_replace7)
    df_train['Q9'] = df_train['Q9'].apply(nan_replace7)
    df_train['Q10'] = df_train['Q10'].apply(nan_replace7)
    df_train['Q11'] = df_train['Q11'].apply(nan_replace7)
    df_train['Q12'] = df_train['Q12'].apply(nan_replace7)
    df_train['Q13'] = df_train['Q13'].apply(nan_replace7)
    df_train['Q14'] = df_train['Q14'].apply(nan_replace7)
    df_train['Q15'] = df_train['Q15'].apply(nan_replace7)
    df_train['Q16'] = df_train['Q16'].apply(nan_replace7)
    df_train['Q17'] = df_train['Q17'].apply(nan_replace9)
    df_train['Q18'] = df_train['Q18'].apply(nan_replace9)
    df_train['Q19'] = df_train['Q19'].apply(nan_replace9)
    df_train['Q20'] = df_train['Q20'].apply(nan_replace9)
    df_train['Q21'] = df_train['Q21'].apply(nan_replace9)
    df_train['Q22'] = df_train['Q22'].apply(nan_replace9)
    df_train['Q23'] = df_train['Q23'].apply(nan_replace9)
    df_train['Q24'] = df_train['Q24'].apply(nan_replace9)
    df_train['Q25'] = df_train['Q25'].apply(nan_replace9)
    df_train['Q26'] = df_train['Q26'].apply(nan_replace9)

    #replace all Nans in marital status and education with unknown values
    df_train['HH_marital'] = df_train['HH_marital'].apply(nan_replace99)
    df_train['HH_education'] = df_train['HH_education'].apply(nan_replace9)

    #replace all Nans in heptatitis_X with 3
    df_train['hepatitis_A_test'] = df_train['hepatitis_A_test'].apply(nan_replace3)
    df_train['hepatitis_B_test'] = df_train['hepatitis_B_test'].apply(nan_replace3)
    df_train['hepatitis_D_test'] = df_train['hepatitis_D_test'].apply(nan_replace3)

    df_train['virusB_infect_status'] = df_train['virusB_infect_status'].apply(nan_replace3)
    df_train['virusC_infect_status'] = df_train['virusC_infect_status'].apply(nan_replace3)
    df_train['virusD_infect_status'] = df_train['virusD_infect_status'].apply(nan_replace3)
    df_train['virusE_infect_status'] = df_train['virusE_infect_status'].apply(nan_replace3)

    df_train['blood_pressure'].fillna(-1, inplace = True)
    df_train['max_inflation_level'].fillna(-1, inplace = True)
    df_train['pulse_status'].fillna(-1, inplace = True)
    df_train['abdomen_diameter'].fillna(-1, inplace = True)
    df_train['weight_status'].fillna(-1, inplace = True)
    df_train['weight'].fillna(-1, inplace = True)    
    df_train['urine_flow_rate'].fillna(-1, inplace = True)
    df_train['glycohemoglobin'].fillna(-1, inplace = True)
    df_train['calcium_level'].fillna(-1, inplace = True)
    df_train['glucose_level'].fillna(-1, inplace = True)
    df_train['iron_level'].fillna(-1, inplace = True)
    df_train['recent_alcohol'].fillna(-1, inplace = True)
    
    df_train= df_train.drop(columns = ['Q27'])
    
    return df_train
```

```python
train_data = preprocessing(df_train)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score
df_test = pd.read_csv("/Users/duytienpham/Downloads/hw3_skeleton/challengeTestUnlabeled.csv")
df_test_X = preprocessing(df_test)
X = train_data.iloc[:,:train_data.shape[1]-1]
y = train_data.iloc[:,-1:]
df_test_X
```

```python
#test
X1 = X.iloc[:1570,:]
X2 = X.iloc[1570:3140,:]
X3 = X.iloc[3140:4710,:]
X4 = X.iloc[4710:6280,:]
X5 = X.iloc[6280:,:]

y1  = y.iloc[:1570,:]
y2  = y.iloc[1570:3140,:]
y3  = y.iloc[3140:4710,:]
y4  = y.iloc[4710:6280,:]
y5  = y.iloc[6280:,:]

#train
X_train1 = (pd.concat([X2,X3,X4,X5])).reset_index().drop(['index'], axis = 1)
X_train2 = (pd.concat([X1,X3,X4,X5])).reset_index().drop(['index'], axis = 1)
X_train3 = (pd.concat([X1,X2,X4,X5])).reset_index().drop(['index'], axis = 1)
X_train4 = (pd.concat([X1,X2,X4,X5])).reset_index().drop(['index'], axis = 1)
X_train5 = (pd.concat([X1,X2,X3,X4])).reset_index().drop(['index'], axis = 1)

y_train1 = (pd.concat([y2,y3,y4,y5])).reset_index().drop(['index'], axis = 1)
y_train2 = (pd.concat([y1,y3,y4,y5])).reset_index().drop(['index'], axis = 1)
y_train3 = (pd.concat([y1,y2,y4,y5])).reset_index().drop(['index'], axis = 1)
y_train4 = (pd.concat([y1,y2,y3,y5])).reset_index().drop(['index'], axis = 1)
y_train5 = (pd.concat([y1,y2,y3,y5])).reset_index().drop(['index'], axis = 1)

X_train = [X_train1,X_train2,X_train3,X_train4,X_train5]
y_train = [y_train1,y_train2,y_train3,y_train4,y_train5]

X_test = [X1,X2,X3,X4,X5]
y_test = [y1,y2,y3,y4,y5]
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
accuracy_rates= []
precision_score1 = []
recall_score1 = []
for i in range(len(X_train)):
    #extract features and labels from training dataframes
  
    clf = AdaBoostClassifier(n_estimators = 50, random_state = 42)
    clf.fit(X_train[i], y_train[i])
    
    train_predicted_y = clf.predict(X_test[i])
    
    accuracy_rates.append(accuracy_score(y_test[i], train_predicted_y))
    
    precision_score1.append(precision_score(y_test[i], train_predicted_y, average = 'macro'))
    recall_score1.append(precision_score(y_test[i], train_predicted_y, average = "macro"))

print (np.mean(accuracy_rates))
print (np.mean(precision_score1))
print (np.mean(recall_score1))
test_predicted_y = clf.predict(df_test_X)
print(test_predicted_y)
```
